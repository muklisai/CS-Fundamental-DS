# NOMOR 2

Read data dari postgres menggunakan python program, convert jadi sebuah dataframe, kemudian save dengan extension .csv file.

````python
# mempersiapkan library
import os
from dotenv import load_dotenv
import psycopg2 as ps
import pandas as pd

# me-laod setting parameter untuk koneksi ke database
load_dotenv()
param = {
  'host' : os.getenv('HOST'),
  'port' : os.getenv('PORT'),
  'user' : os.getenv('USER'),
  'password' : os.getenv('PASS'),
  'database' : os.getenv('DATABASE')
}

# membuat koneksi
conn = ps.connect(**param)

# membuat query
q = 'SELECT * FROM pelukis'

# membaca data dari database
data = pd.read_sql(q, conn)

# mengecek hasil data dari database
display(data)

# menutup koneksi
conn.close()

# export data ke csv
data.to_csv('Data_Pelukis.csv',index=False)
````
