Andi ingin membeli sebuah Laptop seharga Rp $5$ juta. Namun, uang yang dia miliki hanya Rp $3$ juta. Dia teringat bahwa $7$ tahun lalu dia pernah menabung di Bank A dan Bank B dengan saldo yang sama yaitu Rp $1$ juta dan suku bunga yang sama yaitu $6\%$ per tahun. Bank A memberikan suku bunga tunggal dan Bank B memberikan suku bunga majemuk. Jika selama $7$ tahun itu Andi tidak pernah melakukan penarikan maupun pemasukan, Berapakah total uang Andi sekarang ? Apakah uang Andi sudah cukup untuk membeli Laptop ?


```python
uang = 3000000
modal_A = 1000000
modal_B = 1000000
bunga = 0.06
tahun = 7
harga_laptop = 5000000

modal_akhir_A = modal_A * (1 + bunga * tahun)
modal_akhir_B = modal_B * (1 + bunga)**tahun

total_uang = uang + modal_akhir_A + modal_akhir_B
print('total uang Andi: Rp.', format(total_uang, '.2f'))

if total_uang > harga_laptop:
    print('sudah mencukupi')
else: 
    print('tidak mencukupi')
```

## Case 4
Indeks Massa Tubuh (IMT) digunakan untuk menentukan kategori berat badan dengan membandingkan berat dan tinggi badan. Perhitungan IMT dapat dilakukan dengan kalkulator berat badan ideal yaitu : \\
$\text{IMT} = \dfrac{\text{Berat Badan}}{\left(\text{Tinggi Badan}\right)^2}$ \\
Dimana **Berat Badan** dalam **(Kg)** dan **Tinggi Badan** dalam **(meter)**. Untuk populasi Asia, termasuk Indonesia, pengelompokan IMT adalah sebagai berikut : \\
* **Obesitas** = $\text{IMT} \geq 25$
* **Overweight** = $23 \leq \text{IMT} < 25$
* **Normal** = $18,5 \leq \text{IMT} < 23$
* **Underweight** = $\text{IMT} < 18,5$ \\


---


Jika Rudi memiliki Tinggi Badan $175 \text{ cm}$ dan Berat Badan $75 \text{ Kg}$. Termasuk Golongan manakan Berat Badan Rudi?. Buatlah fungsi Python dengan nama **`IMT`** dan parameter fungsi yaitu **`berat`** dalam **Kg** dan **`tinggi`** dalam **cm** untuk menghitung IMT seperti formula di atas. kemudian gunakan fungsi **`IMT`** untuk menentukan kategori dari Berat Badan Rudi !



```python
def IMT(berat, tinggi):
    indeks_massa_tubuh = berat / tinggi**2
    print('IMT =', indeks_massa_tubuh)

    if indeks_massa_tubuh >= 25:
        print('Obesitas')
    elif indeks_massa_tubuh >= 23 and indeks_massa_tubuh < 25:
        print('Overweight')
    elif indeks_massa_tubuh >= 18 and indeks_massa_tubuh < 23:
        print('Normal')
    else:
        print('Underweight')
```

```python
IMT(75, 1.75)
```

1. Kamu akan membuat sebuah class Mahasiswa dan objek-objeknya, untuk mencetak "Biodata Mahasiswa

```python
class Mahasiswa:
    def __init__(self, nama, umur, jenis_kelamin):
        self.nama = nama
        self.umur = umur
        self.jenis_kelamin = jenis_kelamin
    
    def cetak_biodata(self):
        print("Biodata Mahasiswa")
        print("Nama:", self.nama)
        print("Umur:", self.umur)
        print("Jenis Kelamin:", self.jenis_kelamin)

bubu = Mahasiswa("Bubu", 21, "Perempuan")
bubu.cetak_biodata()
```

2. Kamu akan membuat sebuah class LuasBalok yang memiliki atribut/properti panjang dan lebar serta aksinya yang akan dibuat adalah menghitung luasnya dan objeknya Balok

```python
class LuasBalok:
    def __init__(self, panjang, lebar, tinggi):
        self.p = panjang
        self.l = lebar
        self.t = tinggi
    
    def hitung(self):
        return 2 * ((self.p*self.l) + (self.p*self.t) + (self.t*self.l))

balok = LuasBalok(2,3,5)
balok.hitung()
```
